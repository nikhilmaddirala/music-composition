```python
# Search through 
path = '/Users/nikhilmaddirala/Documents/repos/music-composition/carnatify/carnatic-tripod'
txt_path = '/Users/nikhilmaddirala/Documents/repos/music-composition/carnatify/carnatic-tripod/txt/'

#create txt directory
import os
os.makedirs(txt_path, exist_ok=True)

## Search recursively through path, find all .txt files, and copy to txt_path
import os
import shutil
# Get a list of all files in the source directory recursively
files = []
for r, d, f in os.walk(path):
    for file in f:
        files.append(os.path.join(r, file))

# Loop through all files in the source directory
for file in files:
    print(f"found {file}")
    # Check if the file is a .txt file
    if file.endswith('.txt'):
        # Construct the full file path
        src_file = os.path.join(path, file)
        print(f"found text file {src_file}")
        # Copy the file to the destination directory
        shutil.copy(src_file, txt_path)
```

```python
import re

# Define token types and patterns
TOKEN_SPECIFICATION = [
    ('BRACKET_OPEN', r'\(\(*'),      # One or more '('
    ('BRACKET_CLOSE', r'\)\)*'),     # One or more ')'
    ('MODIFIER', r'[.,;]'),          # Duration modifiers
    ('FREQ_MOD', r'[<>]+'),          # Frequency modifiers
    ('ORNAMENT', r'[\/\\-]+'),       # Slides and separators
    ('NOTE', r'[srgmpdn][ai]?(\d)?\*?'),  # Notes with optional variants
    ('WHITESPACE', r'\s+'),          # Whitespace
    ('COMMENT', r'\[.*?\]'),         # Comments enclosed in []
    ('UNKNOWN', r'.'),               # Any other character
]

# Compile the regex patterns
TOKENS_REGEX = '|'.join('(?P<%s>%s)' % pair for pair in TOKEN_SPECIFICATION)
token_re = re.compile(TOKENS_REGEX, re.IGNORECASE)
```
